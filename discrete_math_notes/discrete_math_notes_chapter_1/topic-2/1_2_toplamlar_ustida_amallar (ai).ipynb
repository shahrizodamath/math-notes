# 1.2. To'plamlar ustida amallar

## To'plamlar birlashmasi (yig'indisi)

$$A \cup B$$

> Mantiqdan analog bo'lgan

$$A \cup B = B \cup A \quad \text{— analog } +$$

## To'plamlar kesishmasi (ko'paytmasi)

$$A \cap B$$

$$A \cap B = B \cap A \quad \text{— analog } \cdot$$

$$A \cap B = \emptyset \quad \text{— kesishmaydigan to'plamlar (aksi: kesishadigan)}$$

## To'plamlar ayirmasi

$$A \setminus B = A - B$$

$A$ to'plamdan $B$ to'plamni ayirish amali qo'llanilib, $A \setminus B$
$(A$ va $B$ to'plamlar ayirmasi$)$ to'plami hosil qilindi.

## $A$ to'plamni $B$ gacha to'ldiruvchisi

$$\overline{A}_B = B \setminus A \quad (A \subseteq B \text{ bo'lsa})$$

## Universal to'plam (universum)

Hech bir to'plamga xos qism to'plam bo'lmagan to'plam $\mathcal{U}$ — nisbiy.

Qaralayotgan barcha to'plamni o'zida xos qism to'plam sifatida saqlagan to'plam.

$A$ ni $\mathcal{U}$ gacha to'ldiruvchi emas, $A$ ning to'ldiruvchisi:
$$\overline{A} \cup A = \mathcal{U}$$

$$[\mathcal{U},\ \emptyset] \rightarrow \text{xosmas qism to'plamlar}$$

## To'plam buleani

To'plamning barcha qism to'plamlarini o'zida saqlagan to'plam:

$$2^A$$

$$|A| = 2 \quad \Rightarrow \quad |2^A| = 4$$

$$A = \{a, b\}$$
$$2^A = \{\{a\},\ \{b\},\ \{a, b\},\ \emptyset\}$$

$\emptyset$ — barcha to'plamning xosmas qism to'plami.

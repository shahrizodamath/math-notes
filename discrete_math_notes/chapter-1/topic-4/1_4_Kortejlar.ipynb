# 1.4. Kortejlar (Кортежлар / Tuples)

---

## Asosiy tushuncha

**To'plam** — tartiblanmagan ob'ektlar yig'indisi:
$$A = \{1, 2\} = \{2, 1\}$$

Lekin: $A = \{1, 1, 2\}$ — ? (to'plamda takror bo'lmaydi)

**Kortej** esa tartiblangan va takrorlanishi mumkin bo'lgan elementlar ketma-ketligidir.

## Kortej ta'rifi

Agar $A_1, A_2, A_3, \ldots, A_n$ to'plamlardan har biridan bittadan element $a_1, a_2, \ldots, a_n$ olinsa, ulardan hosil bo'lgan **tartiblangan** ketma-ketlik:

$$\langle a_1, a_2, \ldots, a_n \rangle = K \quad \text{— kortej (elementar kortej)}$$

| Nom | Belgi | Izoh |
|-----|-------|------|
| Juftlik | $\langle a_1, a_2 \rangle$ | 2-darajali kortej |
| Uchlik | $\langle a_1, a_2, a_3 \rangle$ | 3-darajali kortej |
| n-lik | $\langle a_1, \ldots, a_n \rangle$ | n-darajali kortej |

- **n-darajali kortej** = **n o'rinli kortej**
- $|K|$ — kortej **quvvati** (elementlar soni)

## Kortej xususiyatlari

**Kortejda to'plamdan farqli ravishda:**
- Elementlar **takrorlanishi mumkin**
- Joylashgan **o'rni muhim**

$$K \ni a_1, \quad a_1 \in K$$

**Misol:**  
Agar $A = \{1\}$, u holda:
$$K_1 = \langle 1, 1, 1, 1 \rangle, \quad |K_1| = 4$$
$$K_1 = \langle 1, 1 \rangle, \quad |K_1| = 2$$

Agar $A = \{1, 2\}$, $B = \{3\}$:
$$K_2 = \langle 1, 3 \rangle, \quad K_2 = \langle 2, 3 \rangle, \quad K_2 = \langle 1, 1, 1, 3 \rangle \quad \text{va h.k.}$$

```python
# Kortej tushunchasi — Python misoli
# To'plamda tartib va takror muhim emas
A_set = {1, 2}
B_set = {2, 1}
print("To'plamda: A == B?", A_set == B_set)  # True: tartib muhim emas

# Kortejda tartib muhim
K1 = (1, 2)
K2 = (2, 1)
print("Kortejda: K1 == K2?", K1 == K2)  # False: tartib muhim!

# Kortejda takror mumkin
K3 = (1, 1, 1, 1)
print("Takrorli kortej:", K3, "| quvvati:", len(K3))
```

## Kortej elementlari

Kortejning tashkil etuvchilari **kortej elementlari** deyiladi.  
Kortej elementlariga **kortej komponentalari** yoki **koordinatalari** ham deyiladi.

**Misol — kortejlar taqqoslash:**

$$A = \langle \{a,b\},\ 3,\ 4,\ \{c\},\ \{2,5,6\} \rangle$$
$$B = \langle \{a,b\},\ 3,\ 4,\ \{c\},\ \{5,2,6\} \rangle$$
$$C = \langle \{a,b\},\ \{c\},\ 4,\ 3,\ \{2,5,6\} \rangle$$

$$A = B \qquad A \neq C \qquad B \neq C$$
$$|A| = 5 = |B| = |C|$$

> $A = B$ chunki $\{5,2,6\} = \{2,5,6\}$ — bu to'plam, tartib muhim emas.  
> $A \neq C$ chunki komponentalar **o'rni** farq qiladi.

## Geometrik misol — tekislikdagi kortejlar

OX o'qi bo'yicha: $[2, 6]$, OY o'qi bo'yicha: $[2, 4]$

Har bir $\langle x, y \rangle$ — **koordinata kortejlari**  
Bu yerda cheksiz ko'p kortej bor.

$$\langle x, y \rangle \in [2,6] \times [2,4]$$

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(6, 5))

# Hatch bilan to'rtburchak — [2,6] x [2,4]
rect = patches.Rectangle((2, 2), 4, 2, linewidth=2,
                          edgecolor='navy', facecolor='lightblue',
                          alpha=0.5, hatch='//')
ax.add_patch(rect)

# O'qlar
ax.axhline(0, color='black', linewidth=0.8)
ax.axvline(0, color='black', linewidth=0.8)

# Belgilar
ax.annotate('', xy=(7.5, 0), xytext=(0, 0),
            arrowprops=dict(arrowstyle='->', color='black'))
ax.annotate('', xy=(0, 5.5), xytext=(0, 0),
            arrowprops=dict(arrowstyle='->', color='black'))

ax.text(7.6, 0, '$x$', fontsize=12)
ax.text(0.1, 5.6, '$y$', fontsize=12)

# Nuqtalar
for x in [2, 6]:
    ax.plot([x, x], [0, 0.1], 'k-', lw=1)
    ax.text(x - 0.1, -0.3, str(x), fontsize=11)
for y in [2, 4]:
    ax.plot([0, 0.1], [y, y], 'k-', lw=1)
    ax.text(-0.4, y - 0.1, str(y), fontsize=11)

ax.text(3.5, 2.8, r'$\langle x, y \rangle$', fontsize=13, color='navy', ha='center')
ax.set_xlim(-0.5, 8)
ax.set_ylim(-0.5, 6)
ax.set_aspect('equal')
ax.axis('off')
ax.set_title(r'$[2,6] \times [2,4]$ — cheksiz ko\'p kortej', fontsize=12)
plt.tight_layout()
plt.savefig('kortej_geometrik.png', dpi=150)
plt.show()
print("Rasm saqlandi.")
```

## Jadval va kortej

**Misol — O'quvchilar-1 jadvali:**

| No | Ism-familiya | Sinf | Otasi ismi |
|----|-------------|------|------------|
| 1  | Nurmatova Aziza | 5 | Sanjar |
| 2  | Botirov Rasul   | 5 | Akmal  |
| 3  | Azimova Lola    | 5 | Nurbek |

Jadvalda **satrlar ham kortej**:
$$\langle \text{tartib raqam},\ \text{ism},\ \text{sinf},\ \text{otasi ismi} \rangle = K, \quad |K| = 4$$

1 ta satr = 1 ta kortej, 1 ta katakni komponenta.

```python
import pandas as pd

# Jadval — munosabat sifatida (kortejlar to'plami)
oquvchilar_1 = [
    (1, "Nurmatova Aziza", 5, "Sanjar"),
    (2, "Botirov Rasul",   5, "Akmal"),
    (3, "Azimova Lola",    5, "Nurbek"),
]

df1 = pd.DataFrame(oquvchilar_1, columns=["No", "Ism-familiya", "Sinf", "Otasi ismi"])
print("O'quvchilar-1:")
print(df1.to_string(index=False))
print(f"\nHar bir satr = 1 ta kortej, |K| = {len(df1.columns)}")
```

## Dekart ko'paytma

$A_1, A_2, \ldots, A_n$ to'plamlarning **n o'rinli kortejlari to'plami** — **dekart ko'paytma** (to'g'ri ko'paytma) deyiladi.

$$A = A_1 \times A_2 \times A_3 \times \cdots \times A_n = \prod_{i=1}^{n} A_i$$

Agar $A_1 = A_2 = \cdots = A_n = A$ bo'lsa:
$$A_1 \times A_2 \times \cdots \times A_n = A^n \quad \text{— n-darajali dekart ko'paytma}$$

$$\boxed{|A^n| = |A|^n}$$

```python
from itertools import product

A1 = {1, 2}
A2 = {'a', 'b', 'c'}
A3 = {3, 'a'}

# Dekart ko'paytma
cart = list(product(sorted(A1), sorted(A2), sorted(A3, key=str)))

print(f"A1 = {A1}, |A1| = {len(A1)}")
print(f"A2 = {A2}, |A2| = {len(A2)}")
print(f"A3 = {A3}, |A3| = {len(A3)}")
print(f"\n|A1 × A2 × A3| = {len(A1)} × {len(A2)} × {len(A3)} = {len(cart)}")
print("\nA = A1 × A2 × A3:")
for t in cart:
    print(f"  {t}")
```

## Munosabat

**Munosabat** — bu dekart ko'paytmaning **qismi** (n-ar munosabat):

$$R \subseteq A_1 \times A_2 \times \cdots \times A_n$$

**Misol:**  
O'quvchilar-1 va O'quvchilar-2 jadvallarini birlashtirsa ($\cup$) — **5 ta kortejli** munosabat hosil bo'ladi.

```python
# Munosabat = kortejlar to'plami (jadval)
oquvchilar_2 = [
    (1, "Alimov Nodir",   5, "Odil"),
    (2, "Botirov Rasul",  5, "Akmal"),
    (3, "Malikova Ozoda", 5, "Sardor"),
]

# set() — takrorlanmaslik uchun
R1 = set(oquvchilar_1)
R2 = set(oquvchilar_2)

R_union = R1 | R2   # Birlashma (UNION)

df_union = pd.DataFrame(sorted(R_union), columns=["No", "Ism-familiya", "Sinf", "Otasi ismi"])
print("O'quvchilar-1 ∪ O'quvchilar-2:")
print(df_union.to_string(index=False))
print(f"\nUmumiy kortejlar soni: {len(R_union)}")
```

## Excel va kortej

Excelda **jadval nomi** o'zi kortej:

$$\langle \text{jadval nomi},\ \text{satr},\ \text{ustun} \rangle \to \text{manzil kortej}$$

Kortejlardan ko'p joyda foydalaniladi — **yillashtirish raqami ham kortej**.

## Dekart ko'paytmalar birlashmasi

Agar:
$$A = A_1 \times A_2 \times \cdots \times A_n, \qquad B = B_1 \times B_2 \times \cdots \times B_n$$

### Birlashma (≤ munosabati)
$$\prod_{i=1}^{n} A_i \cup \prod_{i=1}^{n} B_i \subseteq \prod_{i=1}^{n} (A_i \cup B_i)$$

### Kesishma (= tenglik)
$$\prod_{i=1}^{n} A_i \cap \prod_{i=1}^{n} B_i = \prod_{i=1}^{n} (A_i \cap B_i)$$

### Bo'shlik sharti
Agar $A_1, A_2, \ldots, A_n$ lardan birortasi $\emptyset$ bo'lsa:
$$\prod_{i=1}^{n} A_i = \emptyset$$

## Katta misol — birlashma va kesishma

$$A = \{1,2\} \times \{a,b,c\} \times \{3,a\}$$
$$B = \{b,c,e\} \times \{a,c,3\} \times \{1,3\}$$

$|A| = 2 \times 3 \times 2 = 12$, $\quad |B| = 3 \times 3 \times 2 = 18$

$$\prod_{i=1}^{3}(A_i \cup B_i) = \{1,2,b,c,e\} \times \{a,b,c,3\} \times \{3,a,1\}$$
$$\left|\prod_{i=1}^{3}(A_i \cup B_i)\right| = 5 \times 4 \times 3 = 60$$

$$|A| + |B| \leq 12 + 18 = 30 \leq 60 \quad \checkmark$$

### Kesishma:
$$A_1 \cap B_1 = \{1,2\} \cap \{b,c,e\} = \emptyset$$
$$\prod_{i=1}^{3}(A_i \cap B_i) = \emptyset \times \{a,c\} \times \{3\} = \emptyset$$

```python
# Katta misol hisoblash
A1 = {1, 2};          B1 = {'b', 'c', 'e'}
A2 = {'a', 'b', 'c'}; B2 = {'a', 'c', 3}
A3 = {3, 'a'};        B3 = {1, 3}

# Birlashma
U1 = A1 | B1; U2 = A2 | B2; U3 = A3 | B3
union_product = list(product(sorted(U1, key=str),
                             sorted(U2, key=str),
                             sorted(U3, key=str)))
print(f"|A| = {len(A1)*len(A2)*len(A3)}, |B| = {len(B1)*len(B2)*len(B3)}")
print(f"|A| + |B| = {len(A1)*len(A2)*len(A3) + len(B1)*len(B2)*len(B3)}")
print(f"|∏(Ai∪Bi)| = {len(U1)} × {len(U2)} × {len(U3)} = {len(union_product)}")

# Kesishma
I1 = A1 & B1; I2 = A2 & B2; I3 = A3 & B3
print(f"\nA1 ∩ B1 = {I1}")
print(f"A2 ∩ B2 = {I2}")
print(f"A3 ∩ B3 = {I3}")
if not I1 or not I2 or not I3:
    print("∏(Ai∩Bi) = ∅  (chunki birortasi bo'sh)")
else:
    inter = list(product(I1, I2, I3))
    print(f"|∏(Ai∩Bi)| = {len(inter)}")
```

## Dekart ko'paytmada kiritish formulasi

$$\prod_{i=1}^{n} A_i \cap \prod_{i=1}^{n} B_i = A \cap B = \prod_{i=1}^{n}(A_i \cap B_i)$$

**Misol bilan tasdiqlash:**
$$= \emptyset \times \{a,c\} \times \{3\} = \emptyset$$

## Dekart ko'paytmaning qism-to'plam xossasi

Agar $A_1 \subseteq B_1,\ A_2 \subseteq B_2,\ \ldots,\ A_n \subseteq B_n$ bo'lsa:

$$A_1 \times A_2 \times \cdots \times A_n \subseteq B_1 \times B_2 \times \cdots \times B_n$$

> Ya'ni **A** dekart ko'paytma **B** dekart ko'paytmaning **qismi** bo'ladi.

```python
# Qism-to'plam xossasini tekshirish
A1_sub = {1}       # ⊆ {1, 2}
A2_sub = {'a'}     # ⊆ {'a', 'b', 'c'}
A3_sub = {3}       # ⊆ {3, 'a'}

B1_full = {1, 2}
B2_full = {'a', 'b', 'c'}
B3_full = {3, 'a'}

cart_A = set(product(A1_sub, A2_sub, A3_sub))
cart_B = set(product(B1_full, B2_full, B3_full))

print(f"A1⊆B1: {A1_sub <= B1_full}")
print(f"A2⊆B2: {A2_sub <= B2_full}")
print(f"A3⊆B3: {A3_sub <= B3_full}")
print(f"\n∏Ai ⊆ ∏Bi: {cart_A <= cart_B}  ✓")
print(f"|∏Ai| = {len(cart_A)}, |∏Bi| = {len(cart_B)}")
```

## Xulosa — Asosiy formulalar

| Formula | Ma'nosi |
|---------|----------|
| $\langle a_1, \ldots, a_n \rangle$ | n-darajali kortej |
| $\prod_{i=1}^{n} A_i$ | Dekart ko'paytma |
| $\|A^n\| = \|A\|^n$ | n-darajali dekart ko'paytma quvvati |
| $\prod A_i \cup \prod B_i \subseteq \prod(A_i \cup B_i)$ | Birlashma (kichiq) |
| $\prod A_i \cap \prod B_i = \prod(A_i \cap B_i)$ | Kesishma (teng) |
| Agar $\exists i: A_i = \emptyset \Rightarrow \prod A_i = \emptyset$ | Bo'shlik sharti |
| Agar $A_i \subseteq B_i\ \forall i \Rightarrow \prod A_i \subseteq \prod B_i$ | Monotonlik |

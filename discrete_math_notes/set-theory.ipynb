## 1. Set Theory

A **set** is a collection of distinct objects.

**Example:** $A = \{1, 2, 3\}$ and $B = \{2, 3, 4\}$

**Union:**
$$A \cup B = \{1, 2, 3, 4\}$$

**Intersection:**
$$A \cap B = \{2, 3\}$$

```python
A = {1, 2, 3}
B = {2, 3, 4}

print("Union:", A | B)
print("Intersection:", A & B)
print("Difference:", A - B)
```

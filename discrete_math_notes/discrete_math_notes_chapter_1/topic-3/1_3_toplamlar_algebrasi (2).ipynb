# 1.3. To'plamlar algebrasi

> **Shart:** $\forall A, B, C, \mathcal{U}$ — $\mathcal{U}$-dan tashqari emas

## Asosiy qonunlar

**① $\emptyset$ ning xossalari:**
$$A \cup \emptyset = A, \quad \emptyset \cup A = A$$
$$A \cap \emptyset = \emptyset, \quad \emptyset \cap A = \emptyset$$

**② $\mathcal{U}$ ning xossalari:**
$$A \cup \mathcal{U} = \mathcal{U}, \quad \mathcal{U} \cup A = \mathcal{U}$$
$$A \cap \mathcal{U} = A, \quad \mathcal{U} \cap A = A$$

**③ Idempotentlik / Involyutivlik:**
$$A \cup A = A, \quad A \cap A = A$$

**④ Nol va birning bog'liqligi:**
$$\overline{\mathcal{U}} = \emptyset, \quad \overline{\emptyset} = \mathcal{U}$$

**⑤ Involyutivlik:**
$$\overline{\overline{A}} = A$$

## Isbot: $A \cup \emptyset = A$

$(\Rightarrow)$ $\forall x \in (A \cup \emptyset) \Rightarrow x \in A$ yoki $x \in \emptyset$

$|\emptyset| = 0 \Rightarrow x \in A$

Ta'rif: $A \cup \emptyset \subseteq A$

$(\Leftarrow)$ $\forall x \in A \Rightarrow x \in A \cup \emptyset$

$A \subseteq A \cup \emptyset$

$$\Rightarrow A \cup \emptyset = A \quad \blacksquare$$

## Isbot: $A \cap \emptyset = \emptyset$

$(\Rightarrow)$ $\forall x \in A \cap \emptyset \Rightarrow x \in A$ va $x \in \emptyset \Rightarrow x \in \emptyset$

$|\emptyset| = 0 \Rightarrow A \cap \emptyset \subseteq \emptyset$

$\emptyset \subseteq A \cap \emptyset$ ($\forall B$ uchun $\emptyset \subseteq B$)

$$A \cap \emptyset = \emptyset \quad \blacksquare$$

**Ziddiyat bilan:** Faraz $A \cap \emptyset = B$ bo'lsin, $B \neq \emptyset$.

$\forall x \in A \cap \emptyset \Rightarrow x \in A$ va $x \in \emptyset \Rightarrow x \in \emptyset$, $|\emptyset|=0 \Rightarrow B = \emptyset$ — ziddiyat. $\blacksquare$

## Isbot: $A \cup A = A$ va $A \cap A = A$

$A \cup A \xrightarrow{\text{birlashma ta'rifi}} A \Rightarrow A \cup A = A$

$A \cap A$ — kesishma ta'rifi $\Rightarrow A \cap A = A \quad \blacksquare$

## Isbot: $\overline{\mathcal{U}} = \emptyset$ va $\overline{\emptyset} = \mathcal{U}$

$\overline{\mathcal{U}} = \mathcal{U} \setminus \mathcal{U} = \emptyset$ — ta'rif

$\overline{\emptyset} = \mathcal{U} \setminus \emptyset = \mathcal{U}$ $\quad \blacksquare$

## Isbot: $\overline{\overline{A}} = A$

$\overline{A} \xrightarrow{\text{to'ld.}} \overline{\overline{A}} \xrightarrow{\text{ning to'ld.}} A$

$(\Rightarrow)$ $\forall x \in \overline{\overline{A}} \Rightarrow x \notin \overline{A} \Rightarrow x \in A \Rightarrow \overline{\overline{A}} \subseteq A$

$(\Leftarrow)$ $\forall x \in A \Rightarrow x \notin \overline{A} \Rightarrow x \in \overline{\overline{A}} \Rightarrow A \subseteq \overline{\overline{A}}$

$$\overline{\overline{A}} = A \quad \blacksquare$$

## Birlashmaga nisbatan kommutativlik

$$A \cup B = B \cup A$$

$\forall x \in A \cup B \Rightarrow x \in A$ yoki $x \in B$
$\Leftrightarrow x \in B$ yoki $x \in A \Rightarrow x \in B \cup A$

$A \cup B \subseteq B \cup A$, xuddi shunday $B \cup A \subseteq A \cup B$

$$A \cup B = B \cup A \quad \blacksquare$$

## Birlashmaga nisbatan assotsiativlik

$$(A \cup B) \cup C = A \cup (B \cup C) \Rightarrow A \cup B \cup C$$

$$A_1 \cup A_2 \cup \ldots \cup A_n = \bigcup_{i=1}^{n} A_i$$

**①** $(A \cup B) \cup C = A \cup B \cup C$ (assotsiativlik)

$$A_i \cup A_{i+1} \Rightarrow (A_i \cup A_{i+1}) \cup A_{i+3} \Rightarrow A_1 \cup A_2 \cup A_3 \cup \ldots \cup A_n = \bigcup_{i=1}^{n} A_i$$

**②** $A \cup B = B \cup A \Rightarrow A_i \cup A_{i+1}$

$$(A_i \cup A_{i+1}) \cup A_{i+3} \Rightarrow A_1 \cup A_2 \cup A_3 \cup \ldots \cup A_n = \bigcup_{i=1}^{n} A_i$$

$$\left|\bigcup_{i=1}^{n} A_i\right| \leq |A_1| + \ldots + |A_n|$$

$(\Rightarrow)$ $\forall x \in (A \cup B) \cup C \Rightarrow x \in (A \cup B)$ yoki $x \in C$
$\Leftrightarrow x \in A$ yoki $x \in B$ yoki $x \in C \Rightarrow x \in A \cup (B \cup C)$

$(A \cup B) \cup C \subseteq A \cup (B \cup C)$; xuddi shunday teskari

$$(A \cup B) \cup C = A \cup (B \cup C) \quad \blacksquare$$

## Kesishmaga nisbatan kommutativlik

$$A \cap B = B \cap A$$

## Kesishmaga nisbatan assotsiativlik

$$(A \cap B) \cap C = A \cap (B \cap C) \Rightarrow A \cap B \cap C$$

$$A_1 \cap A_2 \cap \ldots \cap A_n = \bigcap_{i=1}^{n} A_i$$

$$(A_i \cap A_{i+1}) \cap A_{i+2} \Rightarrow A_1 \cap A_2 \cap \ldots \cap A_n = \bigcap_{i=1}^{n} A_i \quad (\text{kom.})$$

$$\left|\bigcap_{i=1}^{n} A_i\right| \leq \min_{1 \leq i \leq n} |A_i|$$

## Distributivlik (birlashmaga nisbatan)

$$A \cup (B \cap C) = (A \cup B) \cap (A \cup C)$$

$(\Rightarrow)$ $\forall x \in A \cup (B \cap C) \Rightarrow x \in A$ yoki $x \in B$ va $x \in C$

$\Rightarrow x \in A$ yoki $x \in B$ va $x \in A$ yoki $x \in C$

$\Rightarrow x \in (A \cup B) \cap (A \cup C) \supseteq A \cup (B \cap C)$

$(\Leftarrow)$ $\forall x \in (A \cup B) \cap (A \cup C) \Rightarrow x \in A$ yoki $x \in B$ va $x \in A$ yoki $x \in C$

$\Rightarrow x \in A$ yoki $x \in B \cap C \Rightarrow x \in A \cup (B \cap C) \supseteq (A \cup B) \cap (A \cup C)$

$$A \cup (B \cap C) = (A \cup B) \cap (A \cup C) \quad \blacksquare$$

> Eslatma: $3 \times (2 + 4) = 3 \cdot 2 + 3 \cdot 4$

## Distributivlik (kesishmaga nisbatan)

$$A \cap (B \cup C) = (A \cap B) \cup (A \cap C)$$

$(\Rightarrow)$ $\forall x \in A \cap (B \cup C) \Rightarrow x \in A$ va $x \in B$ yoki $x \in C$

$\Rightarrow x \in A$ va $x \in B$ yoki $x \in A$ va $x \in C$

$\Rightarrow x \in (A \cap B) \cup (A \cap C) \supseteq A \cap (B \cup C)$

$(\Leftarrow)$ $\forall x \in (A \cap B) \cup (A \cap C) \Rightarrow x \in A$ va $x \in B$ yoki $x \in A$ va $x \in C$

$\Rightarrow x \in A$ va $x \in B \cup C \Rightarrow x \in A \cap (B \cup C)$

$$A \cap (B \cup C) = (A \cap B) \cup (A \cap C) \quad \blacksquare$$

**Umumiy holat:**
$$A \cap (A_1 \cup A_2 \cup \ldots \cup A_n) = (A \cup A_1) \cap (A \cup A_2) \cap \ldots \cap (A \cup A_n)$$

## De Morgan qonuni (birlashmaga nisbatan)

$$\overline{A \cup B} = \overline{A} \cap \overline{B}$$

Venn diagrammasi (4 holat):

$(\Rightarrow)$ $\forall x \in \overline{A \cup B} \Rightarrow x \notin A \cup B$

$(4)$ $x \in \mathcal{U}$

$\Rightarrow x \notin A$ va $x \notin B$

$\Rightarrow x \in \overline{A}$ va $x \in \overline{B} \Rightarrow x \in \overline{A} \cap \overline{B} \supseteq \overline{A \cup B}$

$(\Leftarrow)$ $\forall x \in \overline{A} \cap \overline{B} \Rightarrow x \notin A$ va $x \notin B \Rightarrow x \notin A \cup B$

$\Rightarrow x \in \overline{A \cup B} \supseteq \overline{A} \cap \overline{B}$

$$\overline{A \cup B} = \overline{A} \cap \overline{B} \quad \blacksquare$$

## De Morgan qonuni (kesishmaga nisbatan)

$$\overline{A \cap B} = \overline{A} \cup \overline{B}$$

$(\Rightarrow)$ $\forall x \in \overline{A \cap B} \Rightarrow x \notin A \cap B$

**(a)** $x \notin A$ yoki **(b)** $x \notin B$ yoki **(d)** $x \notin A$ va $x \notin B$

$\Rightarrow$ **(a)** $x \in \overline{A}$ yoki **(b)** $x \in \overline{B}$ yoki **(d)** $x \in \overline{A}$ va $x \in \overline{B}$

$\Rightarrow x \in \overline{A} \cup \overline{B} \supseteq \overline{A \cap B}$

$(\Leftarrow)$ **(A) ga ko'ra:** $\forall x \in \overline{A} \cup \overline{B} \Rightarrow x \notin A \cap B \Rightarrow x \in \overline{A \cap B} \supseteq \overline{A} \cup \overline{B}$

$$\overline{A \cap B} = \overline{A} \cup \overline{B} \quad \blacksquare$$

## Yutilish qonuni (birlashmaga nisbatan)

$$A \cup (A \cap B) = A$$

$A \cap \mathcal{U} = A$ ($\mathcal{U}$ ning xossati)

$$A \cup (A \cap B) = (A \cap \mathcal{U}) \cup (A \cap B) \xrightarrow{\text{distrib.}} A \cap (\mathcal{U} \cup B) = A \cap \mathcal{U} = A \quad \blacksquare$$

## Yutilish qonuni (kesishmaga nisbatan)

$$A \cap (A \cup B) = A$$

$A = A \cup A$ (idempotentlik)

$$(A \cup A) \cap (A \cup B) = A \cup (A \cap B) = A \quad \blacksquare$$

## O'zaro ko'rinma tengliklar (2 taraflama)

$$\cup \rightarrow \cap, \quad \cap \rightarrow \cup, \quad \emptyset \rightarrow \mathcal{U}, \quad \mathcal{U} \rightarrow \emptyset$$

**Misol:** $A \cup \mathcal{U} = \mathcal{U} \Rightarrow A \cap \emptyset = \emptyset$

## Ekvivalent tasdiqlar

$$A \subseteq B \Leftrightarrow A \cap B = A \Leftrightarrow A \cup B = B$$

**① $A \subseteq B \Rightarrow A \cap B = A$:**

$A \subseteq B$ bo'lsin. $\forall x \in A, A \subseteq B \Rightarrow x \in B$

$x \in A \Rightarrow x \in A \cap B \supseteq A$. $A \cap B \subseteq A$ — har doim.

$A \cap B = A \quad \blacksquare$

**② $A \cap B = A \Rightarrow A \cup B = B$:**

$$A \cup B = (A \cap B) \cup B = B \cup (A \cap B) = B \cup (B \cap A) = B \text{ (yutilish)} \quad \blacksquare$$

**③ $A \cup B = B \Rightarrow A \subseteq B$:**

F.q. $\exists x \in A \Rightarrow x \notin B$

$x \in A \Rightarrow x \in A \cup B = B \Rightarrow x \in B$ — ziddiyat.

Faraz-noto'g'ri. $A \subseteq B \quad \blacksquare$

## Masala 1

**a)** $(A \cap B \cap C) \cup (\overline{A} \cap B \cap C) \cup \overline{B} \cup \overline{C} = \mathcal{U}$

$(A \cap B \cap C) \cup \overline{B} \cup (\overline{A} \cap B \cap C) \cup \overline{C} =$

$= (A \cup \overline{B}) \cap (B \cup \overline{B}) \cap (C \cup \overline{B}) \cup \ldots$

$= (A \cap (B \cap C)) \cup (\overline{A} \cap (B \cap C)) \cup \overline{B} \cup \overline{C}$

$= (A \cup \overline{A}) \cap (B \cap C) \cup \overline{B} \cup \overline{C}$

$= (B \cap C) \cup \overline{B} \cup \overline{C}$

$= (B \cup \overline{B}) \cap (C \cup \overline{B}) \cup \overline{C}$

$= \mathcal{U} \cap (C \cup \overline{B}) \cup \overline{C} = (C \cup \overline{B}) \cup \overline{C} = (C \cup \overline{C}) \cup \overline{B} = \mathcal{U} \quad \blacksquare$

**b)** $(A \cap B \cap C \cap \overline{X}) \cup (\overline{A} \cap C) \cup (\overline{B} \cap C) \cup (C \cap X) = C$

$((A \cap B) \cap (C \cap \overline{X})) \cup ((\overline{A} \cup \overline{B}) \cap C) \cup (C \cap X)$

$= ((A \cap B) \cap (C \cap \overline{X})) \cup (\overline{A \cap B} \cap C) \cup (C \cap X)$

$= (((A \cap B) \cup (C \cap X)) \cap ((C \cap \overline{X}) \cup (C \cap X))) \cup (\overline{A \cap B} \cap C)$

$= ((A \cap B) \cup (C \cap X)) \cap C \cup (\overline{A \cap B} \cap C)$

$= (((A \cap B) \cup (C \cap X)) \cup \overline{A \cap B}) \cap C$

$= (\mathcal{U} \cup (C \cap X)) \cap C = \mathcal{U} \cap C = C \quad \blacksquare$

## Masala 2

**a)** $\forall A$ uchun $A \cup B = A \Rightarrow B = \emptyset$

$A = \emptyset \Rightarrow \emptyset \cup B = \emptyset \Rightarrow B = \emptyset \quad \blacksquare$

**b)** $\forall A$ uchun $A \cap B = A \Rightarrow B = \mathcal{U}$

$A \cap B \subseteq A$, $A = \mathcal{U}$: $\mathcal{U} \cap B = \mathcal{U} \Rightarrow B = \mathcal{U} \quad \blacksquare$

**d)** $A \cup B = \mathcal{U}$ va $A \cap B = \emptyset \Rightarrow B = \overline{A}$ va $A = \overline{B}$

$A \cup B = \mathcal{U}$: $\forall y \notin A \Rightarrow y \in B \Rightarrow \overline{A} \subseteq B$

$\forall y \in \mathcal{U} = A \cup B \Rightarrow y \in B$; $\forall y \notin A \Rightarrow y \in \overline{A} \Rightarrow y \in B \Rightarrow \overline{A} \subseteq B$

$A \cap B = \emptyset$: $\forall x \in A \Rightarrow x \notin B$. $B$ ning hech bir elementi $A$ da yotmaydi, demak $\overline{A}$ da yotadi: $B \subseteq \overline{A}$

$B = \overline{A} \quad \blacksquare$

## Masala 3

> $X \setminus Y = X \cap \overline{Y}$

**a)** $A \setminus (B \cup C) = (A \setminus B) \cap (A \setminus C)$

$[X \setminus Y = X \cap \overline{Y}]$

$$A \cap \overline{(B \cup C)} = A \cap (\overline{B} \cap \overline{C}) = (A \cap \overline{B}) \cap (A \cap \overline{C}) \quad \blacksquare$$

**b)** $A \setminus (B \cap C) = (A \setminus B) \cup (A \setminus C)$

$$A \cap \overline{(B \cap C)} = A \cap (\overline{B} \cup \overline{C}) = (A \cap \overline{B}) \cup (A \cap \overline{C}) \quad \blacksquare$$

**d)** $A \setminus B = A \setminus (A \cap B)$

$$A \cap \overline{B} = A \cap \overline{(A \cap B)} = A \cap (\overline{A} \cup \overline{B}) = (A \cap \overline{A}) \cup (A \cap \overline{B}) = \emptyset \cup (A \cap \overline{B}) = A \cap \overline{B} \quad \blacksquare$$

**e)** $A \setminus (A \setminus B) = A \cap B$

$$A \cap \overline{(A \cap \overline{B})} = A \cap (\overline{A} \cup B) = (A \cap \overline{A}) \cup (A \cap B) = \emptyset \cup (A \cap B) = A \cap B \quad \blacksquare$$

**f)** $A \cup B = A \cup (B \setminus A)$

$$A \cup (B \cap \overline{A}) = (A \cup B) \cap (A \cup \overline{A}) = (A \cup B) \cap \mathcal{U} = A \cup B \quad \blacksquare$$

**g)** $(A \setminus B) \setminus C = (A \setminus C) \setminus (B \setminus C)$

$(A \cap \overline{B}) \cap \overline{C} = (A \cap \overline{C}) \cap (\overline{B} \cup C) = A \cap \overline{B} \cap \overline{C} \quad \blacksquare$

**h)** $A \cap (B \setminus C) = (A \cap B) \setminus C$

$A \cap B \cap \overline{C} = A \cap B \cap \overline{C} \quad \checkmark$

**i)** $A \cap (B \setminus C) = (A \cap B) \setminus (A \cap C)$

$$A \cap B \cap \overline{C} = (A \cap B) \cap (\overline{A} \cup \overline{C}) = (A \cap B \cap \overline{A}) \cup (A \cap B \cap \overline{C}) = \emptyset \cup A \cap B \cap \overline{C} \quad \blacksquare$$

**j)** $A \cap (B \setminus A) = \emptyset$

$$A \cap (B \cap \overline{A}) = (A \cap \overline{A}) \cap B = \emptyset \quad \blacksquare$$

**k)** $A \setminus (B \cup C) = (A \setminus B) \setminus C$

$A \cap \overline{B} \cap \overline{C} = A \cap \overline{B} \cap \overline{C} \quad \checkmark$

**l)** $\overline{A \cap \overline{B}} \cup B = \overline{A} \cup B$

$$\overline{A} \cup B \cup B = \overline{A} \cup B \quad \blacksquare$$

**m)** $A \setminus (B \setminus C) = (A \setminus B) \cup (A \cap C)$

$$A \cap (\overline{B} \cup C) = (A \cap \overline{B}) \cup (A \cap C) \quad \blacksquare$$

**n)** $(A \cup B) \setminus C = (A \setminus C) \cup (B \cap C)$

$(A \cup B) \cap \overline{C} = (A \cap \overline{C}) \cup (B \cap \overline{C}) = \overline{C} \cap (A \cup B) \quad \blacksquare$

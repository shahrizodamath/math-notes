## Motivation

In discrete mathematics, we study logical statements and operations:

| Math | Python | Meaning |
|------|--------|---------|
| $A \land B$ | `A and B` | both must be true |
| $A \lor B$ | `A or B` | at least one true |
| $\neg A$ | `not A` | opposite |

In this project, each **sensor** is a logical statement (True/False),  
and we combine them using logic laws to decide: **should the alarm ring?**

---
## Part 1 — The Sensors

Our house has 4 sensors. Each one is either `True` or `False`:

| Variable | Meaning |
|----------|---------|
| $M$ | Motion detected inside the house |
| $D$ | Front door opened |
| $W$ | Window broken |
| $N$ | It is night time |

```python
# --- SENSOR VALUES ---
# Change these to True or False to simulate different situations

M = True   # Motion detected
D = False  # Door opened
W = False  # Window broken
N = True   # Night time

print("=== Current Sensor Status ===")
print(f"Motion detected : {M}")
print(f"Door opened     : {D}")
print(f"Window broken   : {W}")
print(f"Night time      : {N}")
```

---
## Part 2 — Basic Alarm Logic

**Rule:** The alarm triggers if:
- Motion is detected **AND** it is night, **OR**
- The door is opened **AND** it is night, **OR**  
- The window is broken (any time)

In logic notation:
$$\text{ALARM} = (M \land N) \lor (D \land N) \lor W$$

Using the **Distributive Law** $( M \land N) \lor (D \land N) = (M \lor D) \land N$, we can simplify:
$$\text{ALARM} = (M \lor D) \land N \lor W$$

```python
# Original formula
alarm_original = (M and N) or (D and N) or W

# Simplified using Distributive Law
alarm_simplified = ((M or D) and N) or W

print("=== Alarm Decision ===")
print(f"Original formula  : {alarm_original}")
print(f"Simplified formula: {alarm_simplified}")
print()

# Verify they are always equal (Distributive Law proof)
print(f"Are they equal? {alarm_original == alarm_simplified}  ← Distributive Law confirmed!")
print()

if alarm_simplified:
    print("🚨 ALARM IS RINGING!")
else:
    print("✅ Everything is safe. No alarm.")
```

---
## Part 3 — De Morgan's Laws

**De Morgan's Law states:**
$$\neg(A \land B) = \neg A \lor \neg B$$
$$\neg(A \lor B) = \neg A \land \neg B$$

**Application:** When should the alarm be **silent**?

The alarm is silent when it does NOT ring:
$$\text{SILENT} = \neg\text{ALARM} = \neg((M \lor D) \land N \lor W)$$

Applying De Morgan:
$$\text{SILENT} = (\neg M \land \neg D) \lor \neg N) \land \neg W$$

In plain English: **No alarm when (no motion AND no door) OR it's daytime, AND no broken window.**

```python
# De Morgan's Law verification
silent_direct = not alarm_simplified

# Applying De Morgan manually
silent_demorgan = ((not M and not D) or not N) and not W

print("=== De Morgan's Law ===")
print(f"Silent (direct negation) : {silent_direct}")
print(f"Silent (De Morgan)       : {silent_demorgan}")
print(f"Are they equal? {silent_direct == silent_demorgan}  ← De Morgan's Law confirmed!")
```

---
## Part 4 — Involution Law

**Involution Law:** $\neg(\neg A) = A$  
Double negation returns to the original value.

**Application:** If we accidentally negate the alarm signal twice, we still get the correct result.

```python
# Involution Law
double_negation = not (not alarm_simplified)

print("=== Involution Law ===")
print(f"Original alarm        : {alarm_simplified}")
print(f"Double negation       : {double_negation}")
print(f"Are they equal? {alarm_simplified == double_negation}  ← Involution Law confirmed!")
```

## Part 5 — Full Simulation

Let's test all possible night scenarios automatically:

```python
print("=== Full Night Simulation ===")
print(f"{'Motion':^8} {'Door':^6} {'Window':^8} {'Night':^6} | {'ALARM':^8}")
print("-" * 45)

for M in [True, False]:
    for D in [True, False]:
        for W in [True, False]:
            for N in [True, False]:
                alarm = ((M or D) and N) or W
                status = "🚨 RING" if alarm else "✅ safe"
                print(f"{str(M):^8} {str(D):^6} {str(W):^8} {str(N):^6} | {status}")
```

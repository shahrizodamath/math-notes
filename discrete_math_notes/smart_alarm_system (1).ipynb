## Motivation

In discrete mathematics, we study logical statements and operations:

| Math | Python | Meaning |
|------|--------|---------|
| $A \land B$ | `A and B` | both must be true |
| $A \lor B$ | `A or B` | at least one true |
| $\neg A$ | `not A` | opposite |

In this project, each **sensor** is a logical statement (True/False),  
and we combine them using logic laws to decide: **should the alarm ring?**

## Part 1 — The Sensors

Our house has 4 sensors:

| Variable | Meaning |
|----------|---------|
| $M$ | Motion detected inside the house |
| $D$ | Front door opened |
| $W$ | Window broken |
| $N$ | It is night time |

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

# --- SENSOR VALUES ---
# Change these to True or False to simulate different situations
M = True   # Motion detected
D = False  # Door opened
W = False  # Window broken
N = True   # Night time

print('=== Current Sensor Status ===')
print(f'Motion detected : {M}')
print(f'Door opened     : {D}')
print(f'Window broken   : {W}')
print(f'Night time      : {N}')
```

## Part 2 — House Visualization

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 7))
bg_color = '#1a1a2e' if N else '#87CEEB'
fig.patch.set_facecolor(bg_color)
ax.set_facecolor(bg_color)

if N:
    for _ in range(50):
        sx, sy = np.random.uniform(0,10), np.random.uniform(5,9)
        ax.plot(sx, sy, '*', color='white', markersize=np.random.uniform(1,4))

ground = plt.Rectangle((0,0), 10, 2, color='#2d5a1b')
ax.add_patch(ground)
house = plt.Rectangle((2,2), 6, 4, color='#c8a882', linewidth=2, edgecolor='#8B6914')
ax.add_patch(house)
roof = plt.Polygon([[1.5,6],[5,8.2],[8.5,6]], color='#8B0000', linewidth=2, edgecolor='#4a0000')
ax.add_patch(roof)

door_color = '#ff4444' if D else '#8B4513'
door = plt.Rectangle((4.2,2), 1.6, 2.2, color=door_color, linewidth=2, edgecolor='black')
ax.add_patch(door)
ax.text(5, 3.1, 'DOOR', fontsize=9, ha='center', color='white', fontweight='bold')
if D: ax.text(5, 2.5, 'OPEN!', fontsize=8, ha='center', color='yellow', fontweight='bold')

win_color = '#ff4444' if W else '#87CEEB'
window_l = plt.Rectangle((2.5,4), 1.5, 1.2, color=win_color, linewidth=2, edgecolor='red' if W else 'black')
ax.add_patch(window_l)
ax.text(3.25, 4.6, 'WIN', fontsize=8, ha='center', color='black')
if W: ax.text(3.25, 4.2, 'BROKEN!', fontsize=7, ha='center', color='white', fontweight='bold')

window_r = plt.Rectangle((6,4), 1.5, 1.2, color='#87CEEB', linewidth=2, edgecolor='black')
ax.add_patch(window_r)
ax.text(6.75, 4.6, 'WIN', fontsize=8, ha='center', color='black')

if M:
    for i in range(3):
        arc = mpatches.Arc((5,5), 1.5+i*0.8, 1.5+i*0.8, angle=0, theta1=0, theta2=360,
                           color='yellow', linewidth=1.5, alpha=0.7-i*0.2)
        ax.add_patch(arc)
    ax.text(5, 3.8, 'MOTION!', fontsize=9, ha='center', color='yellow', fontweight='bold')

alarm = ((M or D) and N) or W
alarm_text = '   ALARM RINGING!   ' if alarm else '   SAFE   '
alarm_color = '#ff0000' if alarm else '#00cc44'
alarm_box = plt.Rectangle((1,0.1), 8, 1.2, color=alarm_color, alpha=0.9, linewidth=2, edgecolor='white')
ax.add_patch(alarm_box)
ax.text(5, 0.75, alarm_text, fontsize=16, ha='center', va='center', color='white', fontweight='bold')

sensors = [('M: Motion', M), ('D: Door', D), ('W: Window', W), ('N: Night', N)]
for i, (name, val) in enumerate(sensors):
    x = 1 + i * 2.3
    color = '#00ff88' if val else '#ff4444'
    box = plt.Rectangle((x-0.3, 9.0), 2, 0.55, color=color, alpha=0.8)
    ax.add_patch(box)
    ax.text(x+0.7, 9.27, f'{name}: {"T" if val else "F"}', fontsize=9,
            ha='center', va='center', color='black', fontweight='bold')

ax.set_xlim(0,10)
ax.set_ylim(0,10)
ax.set_aspect('equal')
ax.axis('off')
ax.set_title('Smart Home Alarm System', fontsize=14, color='white', pad=15, fontweight='bold')
plt.tight_layout()
plt.show()
```

## Part 3 — Alarm Logic & Distributive Law

$$\text{ALARM} = (M \land N) \lor (D \land N) \lor W = (M \lor D) \land N \lor W$$

```python
alarm_original = (M and N) or (D and N) or W
alarm_simplified = ((M or D) and N) or W
print(f'Original  : {alarm_original}')
print(f'Simplified: {alarm_simplified}')
print(f'Equal? {alarm_original == alarm_simplified}  <- Distributive Law confirmed!')
print()
print('ALARM RINGING!' if alarm_simplified else 'Safe. No alarm.')
```

---
## Part 4 — Truth Table (All 16 Combinations)

```python
results = []
for m in [True, False]:
    for d in [True, False]:
        for w in [True, False]:
            for n in [True, False]:
                alarm = ((m or d) and n) or w
                results.append((m, d, w, n, alarm))

fig, ax = plt.subplots(figsize=(10, 8))
headers = ['M (Motion)', 'D (Door)', 'W (Window)', 'N (Night)', 'ALARM']
cell_data = [['T' if v else 'F' for v in row] for row in results]
cell_colors = []
for row in results:
    colors = []
    for i, val in enumerate(row):
        if i == 4:
            colors.append('#ff6b6b' if val else '#6bcb77')
        else:
            colors.append('#fff3b0' if val else '#e8e8e8')
    cell_colors.append(colors)

table = ax.table(cellText=cell_data, colLabels=headers,
                 cellLoc='center', loc='center', cellColours=cell_colors)
table.auto_set_font_size(False)
table.set_fontsize(11)
table.scale(1.2, 1.8)
for j in range(5):
    table[0,j].set_facecolor('#4a90d9' if j < 4 else '#cc3300')
    table[0,j].set_text_props(color='white', fontweight='bold')
ax.axis('off')
alarm_count = sum(1 for r in results if r[4])
ax.set_title(f'Truth Table — 16 Combinations\n{alarm_count} alarm situations (red) | {16-alarm_count} safe (green)',
             fontsize=12, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()
```

---
## Part 5 — De Morgan's & Involution Laws

```python
silent_direct = not alarm_simplified
silent_demorgan = ((not M and not D) or not N) and not W
print(f'De Morgan: {silent_direct == silent_demorgan}  <- confirmed!')

double_neg = not (not alarm_simplified)
print(f'Involution: {alarm_simplified == double_neg}  <- confirmed!')
```

## Summary

| Law | Formula | Used for |
|-----|---------|----------|
| **Distributive** | $(M \land N) \lor (D \land N) = (M \lor D) \land N$ | Simplified alarm |
| **De Morgan** | $\neg(A \land B) = \neg A \lor \neg B$ | Silent condition |
| **Involution** | $\neg(\neg A) = A$ | Double negation |

> Boolean logic inside every `if` statement = propositional logic from discrete math. 🎯

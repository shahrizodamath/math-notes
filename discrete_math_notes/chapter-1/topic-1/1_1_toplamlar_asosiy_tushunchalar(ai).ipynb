# 1.1. To'plamlar nazariyasining asosiy tushunchalari

## To'plam va element

To'plamni tashkil etuvchilar shu to'plamning **elementlari** deb ataladi.

- Elementlar soni **chekli** → **chekli to'plam**
- Elementlar soni **cheksiz** → **cheksiz to'plam**

> ⚠️ To'plam elementlari **takrorlanmaydi**. Masalan: $A = \{1, 1, 2\}$ — bu noto'g'ri yozuv.

## To'plamni belgilash usullari

**1. Ro'yxat usuli:**
$$A = \{a, b, c, d, \ldots, z\}$$

**2. Xususiyat usuli (set-builder notation):**
$$B = \{x \mid x = 2n - 1,\ n \in \mathbb{N}\} \quad \text{— toq natural sonlar to'plami}$$

**Tegishlilik belgisi:**
$$a \in A \quad (\text{a element } A \text{ ga tegishli})$$
$$a \notin A \quad (\text{a element } A \text{ ga tegishli emas})$$

## To'plam quvvati (kuchlanishi)

**2-ta'rif.** Chekli to'plamning elementlari soni shu to'plamning **quvvati** deb ataladi.

$$|A| = n$$

**1-misol:**
$$A = \{a\},\quad B = \{a,b\},\quad C = \{a,b,c,d,e\},\quad D = \{1,2,3,\ldots,n\}$$
$$E = \{m \mid m = 2z\},\quad F = \{2,3,5,7,\ldots,p,\ldots\}$$

$A, B, C, D$ — chekli to'plamlar; $E, F$ — cheksiz to'plamlar.

$$|A| = 1,\quad |B| = 2,\quad |C| = 5,\quad |D| = n$$

## Rassel paradoksi va ZFC

**Kantorning sodda to'plamlar nazariyasini** Rassel paradoksi buzdi:

- **Kantor bo'yicha:** $R = \{X \mid X \notin X\}$ — bu to'plam mavjud
- **ZFC bo'yicha:** $R = \{x \in A \mid x \notin x\}$ — faqat mavjud to'plam ichidan qurish mumkin

**Zermelo–Frenkel (+ Tanlash aksiomasi) → ZFC**

Bundan boshqa aksiomatik nazariyalar ham mavjud: von Neyman–Berneys–Gödel (NBG) va h.k.

## ZFC aksiomalari (asosiy)

**Hajmiylik aksiomasi:**
$$A = B \iff (A \subseteq B\ \text{ va }\ B \subseteq A)$$

**Bo'sh to'plam aksiomasi:**

Birorta ham elementga ega bo'lmagan to'plam — **bo'sh to'plam** — mavjud.
$$\emptyset \subset A \quad \text{(har qanday } A \text{ uchun)}$$
$$|\emptyset| = 0, \quad |\{\emptyset\}| = 1$$

**Juftlik aksiomasi:**

Ixtiyoriy $A$ va $B$ to'plamlar uchun $\exists\, C$:
$$C = \{A, B\}, \quad |C| = 2$$
$$\text{Agar } A = B \Rightarrow C = \{A\} = \{B\},\quad |C| = 1$$

**Tanlash aksiomasi:**

Kesishmaydigan to'plamlar majmuasidan $\exists\, X_1, X_2, \ldots, X_n$ — har biridan bittadan element tanlab:
$$C = \{x_1, x_2, \ldots, x_n\}$$
$C$ va $X_i$ lar faqat bitta umumiy elementga ega.

> Bu aksiomasiz cheksiz to'plamlar uchun juda qiziq paradokslar (Banax–Tarskiy, Hausdorff) kelib chiqadi.

## Qism to'plam va xos qism to'plam

**3-ta'rif (Qism to'plam):**
Agar $B$ to'plamning har bir elementi $A$ to'plamda ham mavjud bo'lsa:
$$B \subseteq A \quad \text{yoki} \quad A \supseteq B$$

$$A \subseteq B\ \text{ va }\ B \subseteq A \implies A = B$$

Har qanday to'plam o'zi o'zining qism to'plami — bu **refleksivlik** xossasi:
$$A \subseteq A$$

**4-ta'rif (Xos qism to'plam):**
$$B \subset A \iff B \subseteq A\ \text{ va }\ B \neq A$$
$$A \not\subset A \quad \text{(to'plam o'zining xos qismi emas)}$$
$$\emptyset \subset A \quad \text{(bo'sh to'plam har qanday to'plamning xos qismidir)}$$

## Ekvivalent tasdiqlar

**5-ta'rif:** Agar $a \Rightarrow b$ va $b \Rightarrow a$ bo'lsa, $a$ va $b$ tasdiqlar **o'zaro ekvivalent** deb ataladi:
$$a \iff b$$

**Misol:**
$$A \subseteq B,\ B \supseteq A \implies A = B$$
$$A \subseteq B,\ A \supseteq B,\ A = B \quad \text{— bular o'zaro ekvivalent}$$

**2-misol:** $\mathbb{N} \subseteq \mathbb{R}$ — natural sonlar to'plami haqiqiy sonlar to'plamining qism to'plami.

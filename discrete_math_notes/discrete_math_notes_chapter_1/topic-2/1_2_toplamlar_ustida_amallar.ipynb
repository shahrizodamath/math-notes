# 1.2. To'plamlar ustida amallar

## Birlashtirish (Yig'indi)

**Ta'rif.** $A$ va $B$ to'plamlarning **birlashmasi** — ikkala to'plamga tegishli barcha elementlar to'plami:

$$A \cup B = \{x \mid x \in A \ \text{ yoki } \ x \in B\}$$

> Mantiqdan analog: **YOKI** ($\lor$)

**Xossasi (kommutativlik):**
$$A \cup B = B \cup A \quad \text{(qo'shishga analog: } a + b = b + a\text{)}$$

## Kesishma (Ko'paytma)

**Ta'rif.** $A$ va $B$ to'plamlarning **kesishmasi** — ikkala to'plamga ham tegishli elementlar to'plami:

$$A \cap B = \{x \mid x \in A \ \text{ va } \ x \in B\}$$

> Mantiqdan analog: **VA** ($\land$)

**Xossasi (kommutativlik):**
$$A \cap B = B \cap A \quad \text{(ko'paytirishga analog: } a \cdot b = b \cdot a\text{)}$$

**Kesishmaydigan to'plamlar:**
$$A \cap B = \emptyset \quad \text{(aksi: kesishadigan to'plamlar)}$$

## Ayirma

**Ta'rif.** $A$ va $B$ to'plamlarning **ayirmasi** — $A$ da bor, lekin $B$ da yo'q elementlar to'plami:

$$A \setminus B = A - B = \{x \mid x \in A \ \text{ va } \ x \notin B\}$$

$A$ to'plamdan $B$ to'plamni ayirish amali qo'llanilib, $A \setminus B$ — $(A$ va $B$ to'plamlar ayirmasi$)$ to'plami hosil qilindi.

## To'ldiruvchi

**Ta'rif.** $A$ to'plamni $B$ gacha **to'ldiruvchi** — $B$ da bor, lekin $A$ da yo'q elementlar:

$$\overline{A}_B = B \setminus A \quad (A \subseteq B \text{ bo'lsa})$$

> Ya'ni $B$ to'plamdan $A$ ni ayirish.

## Universal to'plam (Universum)

Hech bir to'lamga xos qism to'plam bo'lmagan to'plam — **universal to'plam** $\mathcal{U}$ → **nisbiy**.

Qaralayotgan barcha to'plamni o'zida xos qism to'plam sifatida saqlagan to'plam:

$$\forall A: \quad A \subseteq \mathcal{U}$$

$A$ ni $\mathcal{U}$ gacha to'ldiruvchi emas — $A$ ning to'ldiruvchisi:
$$\overline{A} \cup A = \mathcal{U}$$

$$[\mathcal{U},\ \emptyset] \rightarrow \text{xosmas qism to'plamlar}$$

## To'plam buleani (Darajali to'plam)

**Ta'rif.** To'plamning barcha qism to'plamlarini o'zida saqlagan to'plam — **bulea** (daraja to'plam):

$$2^A$$

Agar $|A| = n$ bo'lsa:
$$|2^A| = 2^n$$

**Misol:**
$$A = \{a, b\}, \quad |A| = 2, \quad |2^A| = 4$$
$$2^A = \{\{a\},\ \{b\},\ \{a, b\},\ \emptyset\}$$

> $\emptyset$ — barcha to'plamning xosmas qism to'plami.
